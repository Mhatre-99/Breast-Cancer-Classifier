# Visualization

Lets check the target variable first

```python
sns.countplot(cancerData['target'])#0-malignant , 1- benign
```

```python
with pd.option_context('display.max_rows',None,'display.max_columns',None):
    display(cancerData.describe())
```

```python
sns.distplot(cancerData[cancerData['target']==1]['mean radius'])
```

```python
sns.distplot(cancerData[cancerData['target']==0]['mean radius'])
```

We can clearly see that malignant cancer has larger radius than benign cancer

Let's take a look at other characteristics

```python
sns.pairplot(cancerData, vars = ['mean radius',
'mean texture',
'mean perimeter',
'mean area',
'mean smoothness'],
             
hue = 'target')
```

```python
sns.scatterplot(x=cancerData[(cancerData['mean radius']>=13) &(cancerData['mean radius']<=18)]['mean radius'],
                y=cancerData[(cancerData['mean radius']>=13) &(cancerData['mean radius']<=18)]['mean area'],
                hue = cancerData['target'])
```

From the above graphs we can see that it is not necessary that the radius of benign cancer should always be small. There are few exceptions which have radius more than 15. This makes it difficult to identify.

Looking at the pairplot graphs, we can conclude that there is not much of a difference in texture between the two types. So texture alone cannot be used to identify the cancer. And same goes for the smoothness.

Malignant cancers have more variation in their size than Benign cancers. Like we can see the distribution of radius, perimeter and area. This makes it difficult to identify malignant cancers which are smaller in size. Since they can be mistaken for Benign cancer.


```python
sns.scatterplot(x='mean area', y='mean smoothness', hue = 'target', data = cancerData)
```

Even though the smoothness doesn't really help in differentiating between the two cancer types, area on the other hand is very useful in predicting the type. If the area if larger it is most likely to be malignant.
The problem arises when the area is not that big and not that small, when it lies in the middle. 

```python
sns.scatterplot(cancerData['mean concave points'],cancerData['mean radius'], hue = cancerData.target)
```

# Correlation with target variables

```python
corr = cancerData.corrwith(cancerData.target)
```

```python
corr.plot.bar(figsize = (12,12))
```

Okay so the correlation graph proves that our assumptions about radius, perimeter, area, texture, smoothness were true.
As discussed since there is a huge difference between the areas of the two types it is obvious that area will have a higher correlation with the target variable. Same goes for perimeter and radius. Where as texture and smoothness have lesser correlation. 

```python
plt.figure(figsize = (20,10))
sns.heatmap(cancerData.drop(columns=['target']).corr(), annot = True)
```

# Model Training

```python
x = cancerData.iloc[: , :-1]
```

```python
x
```

```python
y = cancerData.iloc[:, -1:]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 5)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
```

```python
regressor = SVC()
```

```python
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
cr = classification_report(y_test, y_pred)
```

```python
print(cr)
```

```python
sns.heatmap(cm, annot = True)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1,10,100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(x_test)
```

```python
cm1 = confusion_matrix(y_test, grid_pred)
```

```python
sns.heatmap(cm1, annot = True)
```

```python
print(classification_report(y_test, grid_pred))
```
